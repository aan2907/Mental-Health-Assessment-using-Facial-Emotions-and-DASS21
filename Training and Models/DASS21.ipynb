```python
questions= ["I found it hard to wind down", "I was aware of dryness of my mouth", "I couldn’t seem to experience any positive feeling at all", "I experienced breathing difficulty (e.g. excessively rapid breathing, breathlessness in the absence of physical exertion)", "I found it difficult to work up the initiative to do things", "I tended to over-react to situations", "I experienced trembling (e.g. in the hands)", "I felt that I was using a lot of nervous energy", "I was worried about situations in which I might panic and make a fool of myself", "I felt that I had nothing to look forward to", "I found myself getting agitated", "I found it difficult to relax", "I felt down-hearted and blue", "I was intolerant of anything that kept me from getting on with what I was doing", "I felt I was close to panic", "I was unable to become enthusiastic about anything", "I felt I wasn’t worth much as a person", "I felt that I was rather touchy", "I was aware of the action of my heart in the absence of physical exertion (e.g. sense of heart rate increase, heart missing a beat)", "I felt scared without any good reason", "I felt that life was meaningless"]
```

```python
len(questions)
```

```python
def evaluate_stress(stress):
  if stress<=14:
    return "Normal"
  elif stress<=18:
    return "Mild"
  elif stress<=25:
    return "Moderate"
  elif stress<=33:
    return "Severe"
  else:
    return "Extreme"
```

```python
def evaluate_anxiety(anxiety):
  if anxiety<=7:
    return "Normal"
  elif anxiety<=9:
    return "Mild"
  elif anxiety<=14:
    return "Moderate"
  elif anxiety<=19:
    return "Severe"
  else:
    return "Extreme"
```

```python
def evaluate_depression(depression):
  if depression<=9:
    return "Normal"
  elif depression<=13:
    return "Mild"
  elif depression<=20:
    return "Moderate"
  elif depression<=27:
    return "Severe"
  else:
    return "Extreme"
```

```python
def assessment(score):
  if score<=14:
    return "Normal", "Focus on self care"
  elif score<=20:
    return "Mild", "Advise to consult a psychologist"
  elif score<=32:
    return "Moderate", "Best to consult a psychologist"
  elif score<=41:
    return "Severe", "Advise to consult a psychiatrist"
  else:
    return "Extreme", "Must see a psychiatrist"
```

```python
stress= [1, 6, 8, 11, 12, 14, 18]
anxiety= [2, 4, 7, 9, 15, 19, 20]
depression= [3, 5, 10, 13, 16, 17, 21]
```

```python
scores= []
print("""0 Did not apply to me at all
1 Applied to me to some degree, or some of the time
2 Applied to me to a considerable degree or a good part of time
3 Applied to me very much or most of the time""")
for ques in questions:
  score= int(input(ques+" 0 1 2 3 :"))
  scores.append(score)

scores
```

```python
stress_score, anxiety_score, depression_score= 0, 0, 0
for i in range(len(scores)):
  if (i+1) in stress:
    stress_score+= scores[i]
  elif (i+1) in anxiety:
    anxiety_score+= scores[i]
  elif (i+1) in depression:
    depression_score+= scores[i]

stress_score*= 2
anxiety_score*= 2
depression_score*= 2

print(stress_score)
print(anxiety_score)
print(depression_score)
```

```python
stress_val= evaluate_stress(stress_score)
anxiety_val= evaluate_anxiety(anxiety_score)
depression_val= evaluate_depression(depression_score)

print(stress_val + " " + anxiety_val + " " + depression_val)
```

```python
total_score= stress_score+anxiety_score+depression_score
total_score/=2
total_score
```

```python
symptom, final_assess= assessment(total_score)
```

```python
print(f"Symptoms are '{symptom}'\nAssessment is '{final_assess}'")
```
