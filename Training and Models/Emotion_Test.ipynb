```python
import torch
import torchvision.transforms as transforms
from efficientnet_pytorch import EfficientNet
from PIL import Image

model = EfficientNet.from_name('efficientnet-b0', num_classes= 7)

model.load_state_dict(torch.load('/content/gdrive/MyDrive/Emotion/emotion_recognition_model.pth'))
model.eval()
```

```python
def preprocess_image(image_path):
    transform = transforms.Compose([
        transforms.Resize((224, 224)),
        transforms.ToTensor(),
        transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225])
    ])
    image = Image.open(image_path).convert('RGB')
    image = transform(image).unsqueeze(0)
    return image

image_path = '/content/gdrive/MyDrive/Emotion/Test/fear/PublicTest_9506885.jpg'
test_image = preprocess_image(image_path)
with torch.no_grad():
    output = model(test_image)

emotion_labels = ["Angry", "Disgust", "Fear", "Happy", "Neutral", "Sad", "Surprise"]

predicted_class_index = torch.argmax(output, dim=1).item()
predicted_emotion = emotion_labels[predicted_class_index]
print("Predicted Emotion:", predicted_emotion)
```
